**Program to get Antonyms from WordNet**

```python
from nltk.corpus import wordnet
```

```python
import nltk
```

```python
pip install nltk
```

```python
nltk.download('wordnet')
```

```python
def get_antonyms(word):
    antonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    return antonyms
```

```python
word = "good"
antonyms = get_antonyms(word)
print(f"Antonyms of {word}: {antonyms}")
```

**Program for stemming non-English words**

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
# Initialize Snowball stemmers for different languages
stemmer_french = SnowballStemmer("french")
stemmer_spanish = SnowballStemmer("spanish")
stemmer_german = SnowballStemmer("german")
stemmer_italian = SnowballStemmer("italian")
stemmer_portuguese = SnowballStemmer("portuguese")
```

```python
# Words for stemming
words = ["mangera", "comiendo", "arbeiten", "mangeranno", "falando"]
```

```python
# Stemming the words for different languages
stemmed_words_french = [stemmer_french.stem(word) for word in words]
stemmed_words_spanish = [stemmer_spanish.stem(word) for word in words]
stemmed_words_german = [stemmer_german.stem(word) for word in words]
stemmed_words_italian = [stemmer_italian.stem(word) for word in words]
stemmed_words_portuguese = [stemmer_portuguese.stem(word) for word in words]
```

```python
# Print the stemmed words for comparison
print("Original Words:", words)
print("French Stemmed Words:", stemmed_words_french)
print("Spanish Stemmed Words:", stemmed_words_spanish)
print("German Stemmed Words:", stemmed_words_german)
print("Italian Stemmed Words:", stemmed_words_italian)
print("Portuguese Stemmed Words:", stemmed_words_portuguese)
```

**Program for lemmatizing words Using WordNet (Use all type of stemmers for the
comparison)**

```python
from nltk.stem import SnowballStemmer, LancasterStemmer, PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# Initialize stemmers
snowball_stemmer = SnowballStemmer("english")
lancaster_stemmer = LancasterStemmer()
porter_stemmer = PorterStemmer()
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Words for lemmatization and stemming
words = ["running", "eating", "flies", "better", "worse"]
```

```python
# Lemmatize and stem the words
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]
snowball_stemmed_words = [snowball_stemmer.stem(word) for word in words]
lancaster_stemmed_words = [lancaster_stemmer.stem(word) for word in words]
porter_stemmed_words = [porter_stemmer.stem(word) for word in words]
```

```python
# Print the original, lemmatized, and stemmed words for comparison
print("Original Words:", words)
print("Lemmatized Words:", lemmatized_words)
print("Snowball Stemmed Words:", snowball_stemmed_words)
print("Lancaster Stemmed Words:", lancaster_stemmed_words)
print("Porter Stemmed Words:", porter_stemmed_words)
```

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# Sample paragraph
paragraph = "The butterfly, a mesmerizing insect known for its delicate beauty and graceful fluttering, undergoes a remarkable transformation from a humble caterpillar. As it emerges from its chrysalis, the stunning wings of the butterfly unfold, revealing a symphony of colors and intricate patterns. These ethereal creatures play a vital role in the ecosystem as pollinators, aiding in the reproduction of various plants. Their ephemeral dance through meadows and gardens captivates the imagination and symbolizes transformation, freedom, and the fleeting nature of life. With their enchanting presence, butterflies bring joy and wonder to people of all ages, inspiring art, literature, and reverence for the natural world."

# Tokenizing the paragraph into words
words = word_tokenize(paragraph)

# Applying PoS tagging
pos_tags = pos_tag(words)

# Applying Named Entity Recognition
named_entities = ne_chunk(pos_tags)

print(named_entities)
```

```python
import spacy

# Load English model for processing
nlp = spacy.load("en_core_web_sm")
```

```python
# The provided paragraph for parsing
paragraph = "The butterfly, a mesmerizing insect known for its delicate beauty and graceful fluttering, undergoes a remarkable transformation from a humble caterpillar. As it emerges from its chrysalis, the stunning wings of the butterfly unfold, revealing a symphony of colors and intricate patterns. These ethereal creatures play a vital role in the ecosystem as pollinators, aiding in the reproduction of various plants. Their ephemeral dance through meadows and gardens captivates the imagination and symbolizes transformation, freedom, and the fleeting nature of life. With their enchanting presence, butterflies bring joy and wonder to people of all ages, inspiring art, literature, and reverence for the natural world."

# Process the paragraph with the NLP pipeline
doc = nlp(paragraph)
```

```python
# Dependency Parsing
print("Dependency Parsing:")
for token in doc:
    print(token.text, "->", token.dep_, "->", token.head.text)

# Constituency Parsing (SpaCy doesn't provide direct support for constituency parsing,
# but we can use displacy to visualize the syntactic structure)
from spacy import displacy

# Create a list of Span objects for the sentence
sent_spans = [doc[:]]

# Visualize the constituency parse tree using displacy
displacy.render(sent_spans, style="dep", jupyter=True, options={"distance": 120})
```
